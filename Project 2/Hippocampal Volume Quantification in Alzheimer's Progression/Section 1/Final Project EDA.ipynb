## Loading NIFTI images using NiBabel

NiBabel is a python library for working with neuro-imaging formats (including NIFTI) that we have used in some of the exercises throughout the course. Our volumes and labels are in NIFTI format, so we will use nibabel to load and inspect them.

NiBabel documentation could be found here: https://nipy.org/nibabel/

Our dataset sits in two directories - *images* and *labels*. Each image is represented by a single file (we are fortunate to have our data converted to NIFTI) and has a corresponding label file which is named the same as the image file.

Note that our dataset is "dirty". There are a few images and labels that are not quite right. They should be quite obvious to notice, though. The dataset contains an equal amount of "correct" volumes and corresponding labels, and you don't need to alter values of any samples in order to get the clean dataset.

```python
# TASK: Your data sits in directory /data/TrainingSet.
# Load an image and a segmentation mask into variables called image and label

dir_training = '/data/TrainingSet'

# Images and labels paths
images_path = os.path.join(dir_training, 'images')
labels_path = os.path.join(dir_training, 'labels')

# All images and labels
images = sorted(glob.glob(os.path.join(images_path, '*')))
labels = sorted(glob.glob(os.path.join(labels_path, '*')))

# Single image and label
image = nib.load(images[0])
label = nib.load(labels[0])

# Getting data
image_data = image.get_fdata()
label_data = label.get_fdata()
```

```python
# Let's see if we have any outliers 
print(len(images), len(labels))
```

We do have an image as an outlier, let's make sure that it is removed

```python
# Get all tails of images directories and store them in a list
images_names = [os.path.split(d)[1] for d in images]
labels_names = [os.path.split(d)[1] for d in labels]

for img in images_names:
    
    # Check the outlier     
    if img not in labels_names:
        print('Outlier: ' + img)
        img = images_path + '/' + img
        images.remove(img)
        
# let's confirm again
print(f'images:{len(images)}, labels: {len(labels)}')
```

```python
# Nibabel can present your image data as a Numpy array by calling the method get_fdata()
# The array will contain a multi-dimensional Numpy array with numerical values representing voxel intensities. 
# In our case, images and labels are 3-dimensional, so get_fdata will return a 3-dimensional array. You can verify this
# by accessing the .shape attribute. What are the dimensions of the input arrays?

# Shapes
print(image_data.shape)
print(label_data.shape)


# TASK: using matplotlib, visualize a few slices from the dataset, along with their labels. 
# You can adjust plot sizes like so if you find them too small:
plt.rcParams["figure.figsize"] = (10,10)


fig, ax = plt.subplots(2,3, figsize = (10, 10))

# Sagital plane
sagittal = image_data[image_data.shape[0]//2,:,:]
sagittal_mask = label_data[image_data.shape[0]//2,:,:]
ax[0,0].imshow(sagittal, cmap="gray")
ax[1,0].imshow(sagittal_mask, cmap="gray")



# Axial plane
axial = image_data[:,:,image_data.shape[2]//2]
axial_mask = label_data[:,:,label_data.shape[2]//2]
ax[0,1].imshow(np.rot90(axial), cmap="gray")
ax[1,1].imshow(np.rot90(axial_mask), cmap="gray")


# Coronal plane
coronal = image_data[:,image_data.shape[1]//2,:]
coronal_mask = label_data[:,image_data.shape[1]//2,:]
ax[0,2].imshow(coronal, cmap="gray")
ax[1,2].imshow(coronal_mask, cmap="gray")
```

## Looking at single image data

```python
# TASK: What dimensions represent axial, sagittal, and coronal slices? How do you know?
print(image.header['dim'])
```

## Plotting some charts

```python
# TASK: Plot a histogram of all volumes that we have in our dataset and see how 
# our dataset measures against a slice of a normal population represented by the chart below.

def volume_cal (index):
    return np.sum(nib.load(labels[index]).get_fdata() != 0)

volume = [volume_cal(i) for i in range(len(labels))]

plt.hist(volume, bins=50)
plt.xlabel('Volume (mmÂ³)');
plt.show()
```

```python
# Let's Investigate one more time but with a different parameter, let's try with the shape
# previous aquired shape: (35, 51, 35)

shapes = [(img, nib.load(img).get_fdata().shape) for img in  images]


threshold_x = int(np.mean([shapes[x][1][0] for x in range(len(shapes))])) * 2
threshold_y = int(np.mean([shapes[x][1][1] for x in range(len(shapes))])) * 2 
threshold_z = int(np.mean([shapes[x][1][2] for x in range(len(shapes))])) * 2

for shape in shapes:
#     print(shape)
    if shape[1][0] > threshold_x or shape[1][1] > threshold_y or shape[1][2] > threshold_z:
        print(f'More Outliers!: {shape}')  
        new_outlier = nib.load(shape[0]).get_fdata()
        
        # Let's plot it
        plt.imshow(new_outlier[:,:,30], cmap='gray')

        
# Okay seems we have two abdominal cross-section images, which are out of our project scope
# Let's remove it, the data path is: '/data/TrainingSet/images/hippocampus_010.nii.gz'

        
images.remove('/data/TrainingSet/images/hippocampus_010.nii.gz')
labels.remove('/data/TrainingSet/labels/hippocampus_010.nii.gz')
                
# let's confirm again
print(f'images:{len(images)}, labels: {len(labels)}')
```
